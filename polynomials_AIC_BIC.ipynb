First use R to get the AIC and BIC of the two models (fixed slope and variable slope) at various polynomial orders, for each subject individually. The following cell will fit these, and export the results into the 'stats' array for plotting by python. (Takes a few dozen seconds on my machine.)

```python
%%R -o stats
df <- read.csv("chung2011_plus.csv", header=TRUE, sep=",")
maxorders <- 40
stats <- array( 0, c(7,maxorders,4) )
for (subject in 1:7)
{
    df1 <- subset( df, sub==sprintf('S%d',subject) ) # extract 1 subject
    for (ord  in 1:maxorders)
    {
        Blocks <- ordered( df1$block )
        Blocks <- C( Blocks, contr.poly, ord)
        
        modl_fixed <- glm(p ~ log10(dur)+Blocks, binomial(probit), data = df1, weights = n)
        modl_var   <- glm(p ~ Blocks + log10(dur)*Blocks, binomial(probit), data = df1, weights = n)
        
        stats[subject,ord,1]=AIC( modl_fixed)
        stats[subject,ord,2]=BIC( modl_fixed)
        stats[subject,ord,3]=AIC( modl_var)
        stats[subject,ord,4]=BIC( modl_var)
    }
}
```

```python
# Python helper routine to plot 1 subject model fitness
def plotsub(stats, subn):
    plt.plot( stats[subn-1,:,0], 'r-', label='Fixed AIC' )
    plt.plot( stats[subn-1,:,1], 'b-', label='Fixed BIC' )
    plt.plot( stats[subn-1,:,2], 'r--', label='Var. AIC' )
    plt.plot( stats[subn-1,:,3], 'b--', label='Var. BIC' )
    plt.xlabel('Polynomial Order')
    plt.ylabel('AIC or BIC')
    plt.text(0.03, 0.97, 'S%d'%(subn), transform=plt.gca().transAxes, va='top', ha='left', size=20) 
    
    plt.gca().set_xscale('log')
    plt.gca().xaxis.set_major_formatter(matplotlib.ticker.ScalarFormatter())
    plt.axis('tight')

    plt.legend( loc='upper right', prop={'size':10})
```

```python
# put all 7 subjects in a grid:
plt.figure( figsize=(16,8))
for sub in np.arange(7):
    plt.subplot(2,4,sub+1)
    plotsub( stats, sub+1)
plt.subplots_adjust( hspace=0.2, wspace=0.3) #fix grid spacing
```
